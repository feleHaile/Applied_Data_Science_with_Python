```python
import datetime
import numpy as np
import pandas as pd
import scipy as sp
import matplotlib as mlt
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
def import_dataframe(filename):
    df = pd.read_excel("./data/Assignment4/"+filename, skiprows = 10)\
            .rename(columns = {"observation_date":"Date", filename[:-4]:filename[1:3] + " Price Index"})\
            .set_index("Date")
    return df
```

```python
def data_plot(df):
    lines = []
    line_colours = ["blue", "green", "red", "orange"]
    line_names = []
    for i in range(len(df.columns.values)):
        if df.columns[i][:2] == "CN":
            line_names.append(df.columns[i][:2] + ".Beijing")
        else:
            line_names.append(df.columns[i][:2])
        line, = plt.plot(df.index,df.iloc[:,i], color = line_colours[i])
        lines.append(line)
    ref_line = plt.plot(df.index, np.full(df.index.shape, 100), color="grey", linestyle=":", alpha = 0.5)
    ax = plt.gca()
    
    ax.set_xlim([datetime.date(2005, 1, 1), datetime.date(2018, 1, 1)])
    ax.set_ylim([60, 160])
    ax.legend(lines, line_names, ncol = 4, frameon = False, loc = 8)
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.set_title("Real Residential Property Prices Index 2005~2017\n(Index 2010=100)")
    plt.tight_layout()
```

```python
rppi_df = \
        import_dataframe("QUSR628BIS.xls").merge(
                import_dataframe("QGBR628BIS.xls").merge(
                        import_dataframe("QCNR628BIS.xls").merge(
                                import_dataframe("QSGR628BIS.xls"), how = "inner", left_index = True, right_index = True), 
                        how = "inner", left_index = True, right_index = True), 
                how = "inner", left_index = True, right_index = True)
```

```python
fig = plt.figure()
data_plot(rppi_df)
```

As we know, the subprime mortgage crisis from 2007 to 2010 (hereinafter referred to as "the crisis") first broke out in the United States, and then spread to the whole world. The crisis was caused by the collapse of a housing bubble, leading to mortgage delinquencies and foreclosures and massive housing price declines. In this visual, the housing price index for the US has drastically dropped from around 155 to 100 between 2006 and 2010, and the drop rate is the highest among the four data sets. Similarly, the price dropping in the United Kingdom and Singapore started in 2007 and stopped after 2009. Although the time and duration of the price dropping were different, the drop rates for those 3 countries during the crisis were almost the same. According to the trend of the red line, it seems that housing prices in Beijing, China were hardly influenced by the crisis. Generatly, the housing price index for Beijing always oscillated in a small range of 85 to 105 from 2005 to 2017. While the housing prices in other 3 countries obviously declined between 2007 and 2009, the index for Beijing slightly increased instead. 

The seemingly abnormal phenomenon mentioned above was a result of a series of financial policies formulated by Chinese government aiming to avoid the impact of the crisis. As a country adopting socialist economy, China may have some drawbacks such as less entrepreneurial opportunity and competition, and a potential lack of motivation by individuals, but when the crisis happened, Chinese government showed its ability and effectiveness to avoid systematic risks and solve domestic problems. All other 3 countries failed to stabilise their economy when the crisis came, and one of the reasons is that the governments doesn't have responsibilities and enough power to intervene in the market to reduce the systematic risks in a very short time.  To summarise, during the subprime mortgage crisis, the socialist economy had higher risk resistance than the capitalism.
